```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('/content/sample_data/bmw_global_sales_2018_2025.csv')
df.dropna()
print(df.head())
```

```python
plt.bar(df.Region,df.Revenue_EUR)
plt.title("AVg price")
plt.show()
```

```python
import seaborn as sns

df_2018 = df[df['Year'] == 2018].sort_values(by='Units_Sold', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x='Model', y='Units_Sold', data=df_2018)
plt.title('Units Sold by Model in 2018 (Sorted)')
plt.xlabel('Model')
plt.ylabel('Units Sold')

plt.show()
```

```python
total =df.groupby('Model')['Units_Sold'].sum().sort_values(ascending=False)
plt.figure(figsize=(8,5))
plt.bar(total.index,total.values)
plt.title("Unit sold")
plt.xlabel('Model')
plt.ylabel('Unit Sold')
plt.show()
```

```python
sns.histplot(data=df,x='Revenue_EUR')
plt.title("Revenue")
plt.xlabel('Dist')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df,x='Model',y='Avg_Price_EUR', palette='Set2')
plt.title('Price')
plt.xlabel('k')
plt.ylabel('f')
plt.show()
```

```python
yearly= df.groupby('Year')['Revenue_EUR'].sum().sort_values(ascending=True)
plt.figure(figsize=(8,5))
plt.plot(yearly.index,yearly.values, marker='o',color='red')
plt.title('Revenue')
plt.xlabel('Year')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='Avg_Price_EUR',y='Units_Sold',hue='Model')
plt.title("Scatter")
plt.xlabel('Avg_Price_EUR')
plt.ylabel('Units_Sold')
plt.show()
```

```python
plt.figure(figsize=(8,5))
new_df=df.select_dtypes(include=['number'])
correlation_matrix = new_df.corr()
sns.heatmap(correlation_matrix,annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

```python
unit=df.groupby('Region')['Units_Sold'].sum()
plt.figure(figsize=(8,8))
plt.pie(unit.values,labels=unit.index,autopct='%1.1f%%')
plt.show()
```
